```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
# Cargar los datos
# Cargar el dataset
df = pd.read_csv('New_BPD.csv')
```

```python
# Mostrar información básica del dataset
print(df.head())
print(df.info())
print(df.describe())
```

```python
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribución de Edad')
plt.show()
```

```python
sns.histplot(data=df, x='Month', kde=True)
plt.title('Distribución de Fechas de Arresto')
plt.show()
```

```python
sns.boxplot(data=df, x='Sex', y='Age')
plt.title('Boxplot de Edad por Sexo')
plt.show()
```

```python
sns.boxplot(data=df, x='Race', y='Age')
plt.title('Boxplot de Edad por Raza')
plt.show()
```

```python
sns.pairplot(df, hue='Race', vars=['Age'])
plt.suptitle('Pairplot de las Características Numéricas del Dataset de df', y=1.02)
plt.show()
```

```python
corr_matrix = df[['Age']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
sns.violinplot(data=df, x='Sex', y='Age')
plt.title('Violin Plot de Edad por Sexo')
plt.show()
```

```python
sns.violinplot(data=df, x='Race', y='Age')
plt.title('Violin Plot de Edad por Raza')
plt.show()
```

```python
sns.countplot(data=df, x='Race')
plt.title('Distribución de las Razas')
plt.show()
```

```python
sns.countplot(data=df, x='Sex')
plt.title('Distribución de los Sexos')
plt.show()
```

```python
sns.lmplot(data=df, x='Age', y='Year', hue='Race', fit_reg=True)
plt.title('Relación entre Edad y Año con Regresión (Dummy)')
plt.show()
```

```python
sns.kdeplot(data=df, x='Age', hue='Race', fill=True)
plt.title('KDE Plot de Edad por Raza')
plt.show()
```

```python

# Creando el gráfico de barras apiladas y normalizadas por raza
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")

# Usando histplot para crear un gráfico de barras apiladas y normalizadas
ax = sns.histplot(
    data=df,
    y='District',
    hue='Race',
    multiple='fill',
    palette='YlGnBu',
    hue_order=df['Race'].value_counts().index
)

# Ajustando la posición de la leyenda fuera del gráfico
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Race')

# Añadiendo etiquetas y título
plt.xlabel('Proporción de Detenciones')
plt.ylabel('Distrito')
plt.title('Proporción de Detenciones por Distrito y Raza')

# Ajustando el diseño para que la leyenda no se superponga al gráfico
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Mostrando el gráfico
plt.show()
```

```python

# Mapa de calor: Arrestados por mes y día de la semana
heatmap_data = df.pivot_table(index='DayOfWeek', columns='Month', aggfunc='size', fill_value=0)
plt.figure(figsize=(14, 5))
ax = sns.heatmap(heatmap_data, cmap='viridis', annot=False, fmt="d", linewidths=.5)
plt.title('Baltimore Arrests by Month and Weekday')
plt.xlabel('Month')
plt.ylabel('Weekday')

# Anotar los valores manualmente
for i in range(heatmap_data.shape[0]):
    for j in range(heatmap_data.shape[1]):
        ax.text(j + 0.5, i + 0.5, heatmap_data.iloc[i, j], 
                ha='center', va='center', color='red', fontsize=8)

plt.show()
```

```python
# Mapa de calor: Arrestados por mes y raza
heatmap_data = df.pivot_table(index='Month', columns='Race', aggfunc='size', fill_value=0)
plt.figure(figsize=(14, 5))
ax = sns.heatmap(heatmap_data, cmap='viridis', annot=False, fmt="d", linewidths=.5)
plt.title('Baltimore Arrests by Month and Race')
plt.xlabel('Race')
plt.ylabel('Month')

# Anotar los valores manualmente
for i in range(heatmap_data.shape[0]):
    for j in range(heatmap_data.shape[1]):
        ax.text(j + 0.5, i + 0.5, heatmap_data.iloc[i, j], 
                ha='center', va='center', color='red', fontsize=8)

plt.show()
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

# Crear la figura y los ejes, especificando el tamaño de la figura
fig, ax = plt.subplots(figsize=(14, 6))

# Plotear el GeoDataFrame en los ejes creados
gdf.plot(column='District', cmap='viridis', legend=True, markersize=5, ax=ax)

# Añadir el título
plt.title('Distribución geográfica de los distritos')

# Mostrar el gráfico
plt.show()
```
